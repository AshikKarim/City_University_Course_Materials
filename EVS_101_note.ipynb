##(b). Illustrate the lentic ecosystem diagram with explanations.---------4

Some examples of lentic ecosystems are lakes, ponds, swamps, marshes, and vernal pools. Lentic ecosystems take many forms, from small, temporary pools to large lakes. Some lentic ecosystems are fresh water with low salt content, and others have a higher salt content (Example: The Great Salt Lake in Utah). Lentic ecosystems such as lakes can be formed by glaciers, volcanoes, and shifting of tectonic plates, and some are man-made. Some, such as vernal pools, are only temporary during a rainy or wet season. Lentic ecosystems have layers from top to bottom that support different organisms, depending on factors such as the amount of light and temperature. Algae and aquatic plants produce food for other organisms in the ecosystem. Many tiny invertebrates, called zooplankton, live in lentic waters. They feed on algae and plants and provide food for other organisms such as snails and insects like water striders. The kinds of fish and other vertebrates depends on many factors, such as the salt content, amount of light, and depth of the water. There are many species of lentic vertebrates, including salamanders, frogs, alligators, and many kinds of birds.

## (c). "Everything that live in an ecosystem is dependent on the other species"- Justify this statement. ------------4

The statement is justified because all organisms in an ecosystem rely on each other for survival. Through food chains, symbiotic relationships, nutrient cycling, habitat sharing, and population control, each species contributes to the balance and health of the ecosystem. If one species is removed, it can disrupt these interactions, impacting others and potentially destabilizing the ecosystem.

## (a). Explain how over exploitation directly create imbalances in biodiversity. ------3

Over-exploitation, such as overfishing, directly harms biodiversity by depleting fish populations, which disrupts food chains and affects entire marine ecosystems. With fewer fish, predator species lose a food source, while prey species might overpopulate, leading to resource shortages and further ecological strain. This imbalance weakens biodiversity, reduces genetic variety, and makes the ecosystem more vulnerable to environmental changes and additional threats.

| **Atmospheric Layer**   | **Location**                        | **Protection Role**                                                                                      |
|-------------------------|-------------------------------------|----------------------------------------------------------------------------------------------------------|
| **Troposphere**         | 0–12 km (closest to Earth's surface) | Contains the majority of Earth’s weather systems, regulates temperature, and supports life.               |
| **Stratosphere**        | 12–50 km                            | Houses the ozone layer, which absorbs and blocks harmful UV radiation from the sun.                        |
| **Mesosphere**          | 50–85 km                            | Burns up most meteoroids and debris entering Earth’s atmosphere, preventing them from reaching the surface.|
| **Thermosphere**        | 85–600 km                           | Absorbs X-rays and extreme UV radiation, protecting life on Earth and supporting the auroras.             |
| **Exosphere**           | 600+ km                             | Outermost layer; gradually fades into space, where particles escape into space, protecting lower layers.   |
